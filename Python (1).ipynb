```python
import csv

#set the floating point precision to 2
%precision 2

#read the csv files and convert to a list of dictionaries
with open('mpg.csv') as csvfile:
    mpg = list(csv.DictReader(csvfile))
    
#the first three dictionaries(elements) in the list
mpg[:3] 
```

```python
#counting how many dictionaries there are in the list
len(mpg)
```

```python
#look for the column names(categories) in the file
mpg[0].keys()
```

```python
#the average city mpg
#string convert to float

sum(float(d['cty']) for d in mpg) / len(mpg)
```

```python
#the average of highway mpg

sum(float(d['hwy']) for d in mpg) / len(mpg)
```

```python
#the average of city mpg grouped by the number of cylinders a car has
#create a set of values of the cylinder to calculate levels for cylinders

cylinders = cylinders = set(d['cyl'] for d in mpg)
cylinders
```

```python
ctympgbycyl = [] #create list to store the calculation

#iterate over all the cylinder levels and the dictionaries
for c in cylinders:
    summpg = 0
    cyltypecount = 0
    for d in mpg:
        if d['cyl'] == c: #if the level matches the level that is calculating for 
            summpg += float(d['cty'])
            cyltypecount += 1 #increment the count
            
    #calculate the average mpg and put it into the list 
    #(append the tuple : ['class', 'avg mpg'])
    ctympgbycyl.append((c, summpg / cyltypecount))
    
#sort from the lowest number of cylinder level to the highest
ctympgbycyl.sort(key = lambda x: x[0])

ctympgbycyl
```

```python
#calculate the average highway mpg for different vehicle classes

#create sets of vehicle classes
vehicleclass = set(d['class'] for d in mpg)
vehicleclass
```

```python
hwympgbyclass = []

for t in vehicleclass:
    summpg = 0
    vclasscount = 0
    for d in mpg:
        if d['class'] == t:
            summpg += float(d['hwy'])
            vclasscount += 1
    hwympgbyclass.append((t, summpg / vclasscount))
    
#sort from the lowest mpg to the highest
hwympgbyclass.sort(key = lambda x: x[1])
hwympgbyclass
```
